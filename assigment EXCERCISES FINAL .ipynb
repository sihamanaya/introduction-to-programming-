```python
# EXCERCISE 1
from datetime import date

# Retrieve the user's birthdate (bd)in the format dd/mm/yyy
bd = input("Enter birthdate in format dd/mm/yyyy: ")

# Divide data entered into month, day, and year

bd_elements = bd.split("/")

# Verify the format and the date entered by the user.

if len(bd_elements) != 3:
   print("Invalid date format, enter the date in dd/mm/yyyy format.")
else:
   day = int(bd_elements[0])
   month = int(bd_elements[1])
   year = int(bd_elements[2])

   if day > 31 or month > 12:
      print("Invalid date entered, enter the date in dd/mm/yyyy format.")
   else:

      # Determine age of user

      today = date.today()
      age = today.year - year - ((today.month, today.day) < (month, day))

      # Determine if user's birthday is today

      if today.month == month and today.day == day:
         print(f"Happy Birthday! You are {age} years old.")
      else:
         print(f"You are {age} years old.")
```

```python
# EXCERCISE 2

import math

# Obtain user input

def get_input():

  # Get numbers from user
  l = int(input("Enter lower range: "))
  h = int(input("Enter higher range: "))

  # Check for invalid input
  if l < 0 or h < 0:

    # Let user know there is an error
    print("Numbers must be positive")

    # Let user retry again
    get_input()

  # Check order and swap if needed
  elif l > h:

    temp = l
    l = h
    h = temp

  return l, h

# Function to determine if a given integer is prime
def prime_numbers (low, high):

  '''# Comment explaining what this function does
  # Takes lower and upper bounds as parameters
  # Returns a list of primes between the two values
'''
  # List to keep only prime numbers
  primes = []

  # Loop over every number in the range.
  for number in range(low, high+1):

  # Until shown otherwise, assume  number is prime.
    confirmed_prime = True

    # Look for components up to the number's square root.
    for i in range(2,int(math.sqrt(number))+2):

      # Not prime if factor divides equally.
      if number % i == 0:

        # Revise prime number value
        confirmed_prime = False

        # End of testing
        break

    # The integer is prime if no factors are identified.
    if confirmed_prime:

      # Add to list of primes
      primes.append(number)

  # Return list of primes
  return primes

# Main function
def main():

  # Get validated input
  l, h = get_input()

  # Call prime function
  primes = prime_numbers(l, h)

  # Print results
  for i in range(0, len(primes), 10):
    print(*primes[i:i+10])

# Call main
main()









```
